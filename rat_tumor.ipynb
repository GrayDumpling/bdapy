# Hierarchical Bayesian Inference of Binomial Probabilities (BDA 5.3) #

This iPython notebook attempts to reproduce plots in Chapter 5.3 of Bayesian Data Analysis (Gelman et al, 3rd Edition).  Here, we are given $J=71$ observations of binomial outcomes $(y_1,N_1), \ldots, (y_J,N_J)$ where $y_j \sim \text{Bin}(N_j, \theta_j)$.  We assume conjugate prior $\theta_j \sim \text{Beta}(\alpha,\beta)$, and a "reasonable" hyperprior density $p(\alpha,\beta) \propto (\alpha+\beta)^{-5/2}$.

```python
# load data from Andrew's homepage
from pandas import *
tumor_df = read_csv("http://www.stat.columbia.edu/~gelman/book/data/rats.asc", skiprows=3, sep=' ')
```

```python
# the data had some blank spaces; remove them
tumor_df = tumor_df[['y','N']]
```

The marginal posterior distribution can be derived as: (equation 5.8)
$$
p(\alpha,\beta \mid y) = \prod_{j=1}^J \frac{\Gamma(\alpha+\beta+n_j)}{\Gamma(\alpha+y_j) \Gamma(\beta+n_j-y_j)}
    \theta_j^{\alpha+y_j-1} (1-\theta_j)^{\beta+n_j-y_j-1}.
$$
Now let us plot the marginal posterior density.  Here we reparametrize hyperparameters to be $\log(\frac \alpha \beta)$ and $\log(\alpha + \beta)$.  As always, when computing a density it is safer to compute its log.

```python
from scipy.special import gammaln
import numpy as np
def log_hyperparam_posterior(df, natural_1, natural_2):
    '''
    log of posterior of hyperparameters eq (5.8) in natural parameter scale 
    ( log(alpha/beta) and log(alpha + beta) )
    '''
    # calculate original parameters
    beta = np.exp(natural_2) / (1.0 + np.exp(natural_1))
    alpha = beta * np.exp(natural_1)
    # now calculate the log density
    log_prior = np.log(alpha) + np.log(beta) - 2.5 * np.log(alpha + beta)
    log_posterior_1 = len(df) * (gammaln(alpha + beta) - gammaln(alpha) - gammaln(beta))
    log_posterior_2 = sum(gammaln(alpha + df['y']) + gammaln(beta + df['N'] - df['y']) - 
            gammaln(alpha + beta + df['N']))
    return log_prior + log_posterior_1 + log_posterior_2
```

```python
import pylab as pl
import numpy as np
n = 64
xmin = -2.3
xmax = -1.3
ymin = 1
ymax = 5
x = np.linspace(xmin, xmax, n)
y = np.linspace(ymin, ymax, n)
X, Y = np.meshgrid(x, y)
log_Z = np.zeros((n,n))
#max_log_Z = float("-inf")
for i in range(n):
    for j in range(n):
        log_Z[i,j] = log_hyperparam_posterior(tumor_df, X[i,j], Y[i,j])
        #max_log_Z = max(max_log_Z, log_Z[i,j])
max_log_Z = log_Z.max()
Z = np.exp(log_Z - max_log_Z) * np.exp(max_log_Z)
levels=np.exp(max_log_Z) * np.linspace(0.05, 0.95, 10)
pl.contourf(X, Y, Z, 8, alpha=.75, cmap='jet', levels=np.exp(max_log_Z) * np.concatenate(([0], np.linspace(0.05, 0.95, 10), [1])))
C = pl.contour(X, Y, Z, 8, colors='black', linewidth=.5, levels=levels)
```

Above is the plot same to Figure 5.3; Figure 5.2 can also be reproduced just by changing the range.

```python
probs = np.ravel(Z)/np.sum(Z)
throws = np.random.choice(len(probs), 1000, p=probs)
xs = x[throws%n]
ys = y[throws/n]
xgrid_size = (xmax-xmin)/n
ygrid_size = (ymax-ymin)/n
xs += numpy.random.random(len(xs)) * xgrid_size - 0.5 * xgrid_size
ys += numpy.random.random(len(ys)) * ygrid_size - 0.5 * ygrid_size
pl.scatter(xs, ys, s=.1)
pl.xlim(xmin, xmax)
pl.ylim(ymin, ymax)
plt.show()
```
